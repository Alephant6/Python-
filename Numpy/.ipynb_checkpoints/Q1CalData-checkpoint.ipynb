```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
from  sklearn  import  datasets
from numpy import random
```

```python
class Cell(object):
    """
    creat class Cell 定义一个格点的类
    """
    nest = False
    T = 1
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "(%d, %d)" %(self.x, self.y)
    
    def print(self):
        print(f"coordinate: ({self.x}, {self.y}), nest: {self.nest}, T: {self.T}")
        
    def setNest(self):
        self.nest = True
        self.T = 0
        
    def array(self):
        return np.array([self.x, self.y])
    
    def distance(self, cell):
        return np.sqrt((self.x-cell.x)**2 + (self.y-cell.y)**2)
        
```

```python
class SetOfCell(list):
    """
    creat class Cell 定义格点的集合类
    """
    def __init__(self, lenOfX, lenOfY):
        list.__init__([])
        self.lenOfX = lenOfX
        self.lenOfY = lenOfY
        for i in range(lenOfX):
            for j in range(lenOfY):
                self.append(Cell(i, j))
    
    def dim(self, x, y):
        if x < self.lenOfX and y < self.lenOfY:
            return self[x*self.lenOfY + y]
        else:
            print("Out of range!")
    
        
        
```

```python
def IdentifyNest(setOfCell):
    """
    区分nest
    """
    setOfNest = []
    for i in range(lenC):
        if setOfCell[i].nest == True:
            setOfNest.append(setOfCell[i])
    return setOfNest
```

```python
def CalDelta(setOfNest, setOfCell):
    """
    计算每个nest到cell的距离
    """
    Delta = np.zeros((lenN, lenC))
    for i in range(lenN):
        for j in range(lenC):
                Delta[i][j] = setOfNest[i].distance(setOfCell[j])
    return Delta
```

```python
def CalC(Delta):
    """
    计算每个nest的竞争系数C
    """
    C = np.exp(-Delta ** 2 / (2 * sigmac ** 2))
    CNest = np.sum(C, axis=1)
    return CNest
```

```python
def CalT(setOfNest):
    T = np.ones(lenC)
    for j in range(lenC):
        if setOfCell[j].nest == True:
            T[j] = 0
    return T
        
```

```python
def CalE(Delta):
    """
    计算每个nest的环境友好指数E
    """
    E = np.zeros((lenN, lenC))
    for i in range(lenN):
        for j in range(lenC):
            E[i][j] = np.exp(-Delta[i][j] ** 2 / (2 * sigmac ** 2)) * T[j]
    ENest = np.sum(E, axis=1) 
    return ENest 
```

```python
def CalCbar():
    return np.sum(CNest) / (lenC) ** 2
```

```python
def CalN(NBefore):
    NLam = r * NBefore / (1 + gamma * CalCbar() * NBefore)
#     print(f"NLam = {NLam}")
    N = random.poisson(lam=NLam)
#     N = NLam
#     print(f"N = {N}")
    return  np.around(N)
    
```

```python
def CalQ(ENest, CNest):
    QLam = np.zeros(lenN)
    QNest = np.zeros(lenN)
    for i in range(lenN):
        QLam[i] = r * ENest[i] / (1 + gamma * CNest[i])
#         print(f"QLam{i} = {QLam[i]}")
        QNest[i] = random.poisson(lam=QLam[i]) * 0.35
#         QNest[i] = QLam[i] * 0.35
#         print(f"QNest{i} = {QNest[i]}")
    QNest = np.around(QNest).astype(int)
    return QNest
    
```

```python
def CalP(Delta):
#     mu = np.mean(Delta, axis=1)
    P = np.zeros((lenN, lenC))
    for i in range(lenN):
        for j in range(lenC):
#             P[i][j] = np.around(np.exp(-Delta[i][j] / mu[i]) * T[j], 3)
            P[i][j] = np.around(np.exp(-Delta[i][j]/(lenOfX/20)) * T[j], 3)
        P[i] = P[i] / np.sum(P[i]) # 对每一个nest的P归一化
#         print(f"P{i} = {P[i]}")
#     P = P / np.sum(P, axis=1).T
    return P
```

```python
def Destination(Q, PNest):
    """设置一个nest的所有queen的着落点
    """
    # 按照概率随机生成目的地
#     np.random.seed(0)
    for i in range(Q):
        index = np.random.choice(range(lenC), p=PNest)
        # 在目的地设置nest
        setOfCell[index].setNest()
    
```

```python
lenOfX, lenOfY = 200, 200
setOfCell = SetOfCell(lenOfX, lenOfY)
lenC = len(setOfCell)
r = 2 # 增长率
gamma = 0.8 # 资源依赖度
T_total = 10
N = np.zeros(T_total)
numNest = np.zeros(T_total)
N[0] = 50
sigmac = 20

# 初始化nest
setOfCell.dim(30,164).setNest()
setOfCell.dim(60,159).setNest()
setOfCell.dim(70,150).setNest()
# for i in range(3):
#     randx = random.randint(lenOfX)
#     randy = random.randint(lenOfY)
#     setOfCell.dim(randx, randy).setNest()


# 更新nest
setOfNest = IdentifyNest(setOfCell)
# nest的个数
lenN = len(setOfNest)
numNest[0] = lenN

CN = []
EN = []
```

```python

for t in range(1, T_total):
    print(t)
    # Calculate Ci and Ei
    # 计算Delta
    Delta = CalDelta(setOfNest, setOfCell)
    # 计算Ci
    CNest = CalC(Delta)
    CN.append(CNest)
    # 计算Ei
    T = CalT(setOfNest)
    ENest = CalE(Delta)
    EN.append(ENest)
    # 计算这一年的黄蜂数Nt
    N[t] = CalN(N[t-1])
    # Randomly select Qi
    # 随机生成queen
    QNest = CalQ(ENest, CNest)
    # Calculate Pij
    P = CalP(Delta)
    for x in range(lenN):
        Destination(QNest[x], P[x])
    # 更新nest
    setOfNest = IdentifyNest(setOfCell)
    # nest的个数
    lenN = len(setOfNest)
    numNest[t] = lenN

        
```

```python
def turnTxTy(T):
    T1 = T.reshape(lenOfX, lenOfY)
    T1x = []
    T1y = []
    for i in range(lenOfX):
        for j in range(lenOfY):
            if T1[i][j] == 0:
                T1x.append(i)
                T1y.append(j)
    return T1x,T1y
# 绘图nest
T = CalT(setOfNest)
Tx, Ty = turnTxTy(T)
fig = plt.figure(figsize=(4,4))
img = plt.imread("Washington.png")
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, 200, 0, 200])
ax.set_aspect('equal', adjustable='box')
ax.axis('off')
ax.scatter(Tx, Ty, s=2, c='r', marker='s')

plt.savefig('./Predicted nest in 2029.jpg', dpi=100)
plt.show()
# plt.imshow(T1,cmap='gray')
```
