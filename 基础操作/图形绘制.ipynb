```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from sklearn import datasets
```

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.plot(x, y*2)
```

```python
plt.plot(x, y, 'y*-')
```

```python
plt.plot(x, y*2, 
         color=colors.to_rgba((0, 0, 0,1)),
        marker='*',
        markersize=5,
        linestyle='--',
        linewidth=10)
plt.title(label='正弦函数')
plt.xlim(0,8)
plt.xlabel(xlabel='x轴')
plt.xticks(ticks=[0,2,4,6,8])
plt.ylim(-3,3)
plt.ylabel(ylabel='y轴')
plt.yticks(ticks=[-3,-2.5,-2,-1.5,-1,-0.5,0,0.5,1,1.5,2,2.5,3])
plt.legend(loc='best')
```

```python
plt.scatter(np.pi,0,s=50)
plt.annotate('flag text',
             xy=(np.pi,0),
             xytext=(2,-2),
             fontsize=24,
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))
```
